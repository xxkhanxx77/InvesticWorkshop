```python
import pandas as pd
import yfinance as yf
```

```python
start = '2021-01-01'
end = '2022-01-01'
```

```python
data = yf.download('TSLA', start=start, end=end)
```

```python
df=data.copy()
```

```python
df_weekly=df.resample('W-FRI').agg({'Open':'first',
                                'High':'max',
                                'Low':'min',
                                'Close':'last',
                                'Volume':'sum'})
```

```python
df_weekly
```

```python
df_monthly=df.resample('M').agg({'Open':'first',
                                'High':'max',
                                'Low':'min',
                                'Close':'last',
                                'Volume':'sum'})
df_monthly
```

```python
df_monthly=df.resample('BM').agg({'Open':'first',
                                'High':'max',
                                'Low':'min',
                                'Close':'last',
                                'Volume':'sum'})
df_monthly
# only business day
```

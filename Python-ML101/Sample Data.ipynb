# let try min max scaler

```python
df_x_normalized = pd.DataFrame(X,columns=data.data.columns)
```

```python
df_x_normalized.describe()
```

```python
upper = df_x_normalized.quantile(.95)
# it mean we cut 5 %`
upper
```

```python
low = .05
high = .95
quant_df = df_x_normalized.quantile([low, high])
# it mean we cut lower 5 % and upper 5%  = 10%
```

```python
quant_df
```

```python
filt_df_x = df_x_normalized.apply(lambda x: x[(x>=quant_df.loc[low,x.name]) &  (x <= quant_df.loc[high,x.name])], axis=0)
```

```python
filt_df_x
```

```python
filt_df_x.dropna()
```

# alternative soltion

```python
Q1 = df_x_normalized.quantile(0.05)
Q3 = df_x_normalized.quantile(0.95)
```

```python
filt_df_x2 = df_x_normalized[~((df_x_normalized < Q1) | (df_x_normalized > Q3 )).any(axis=1)]
filt_df_x2
```

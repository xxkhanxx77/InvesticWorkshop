```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
from sklearn import metrics
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
start='2017-01-01'
end='2021-10-01'
```

```python
data=yf.download('CL=F PTTEP.BK GPSC.BK PTTGC.BK PTT.BK',start=start,end=end)
```

```python
data
```

```python
df=data['Adj Close']
```

```python
df
```

```python
df =df.fillna(method='ffill').pct_change()
```

```python
df=df.dropna()
df
```

```python
df = df[['CL=F','GPSC.BK', 'PTTEP.BK' , 'PTTGC.BK' ,'PTT.BK']]
# re columns in dex put y value to last index
```

```python
df
```

```python
X=df.iloc[:,:-1] # without pttbk
y=df.iloc[:,-1] # only pttbk
```

```python
X,y
```

```python
from sklearn.model_selection import KFold
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

```python
X_train
```

```python
y_train
```

# exmaple Kfold random and result

```python
kfold = KFold(10,shuffle=False,random_state=None)
```

```python
for train,test  in kfold.split(y):
    print('train: %s, test: %s '% (y[train],y[test]))
```

```python
from sklearn.model_selection import cross_val_score
reg= LinearRegression()
scores=cross_val_score(reg,X_train,y_train,scoring='r2',cv=5)
# cv is crossvalidation validator
# cv is K -> very large dataset use , K-3 , K-5 , normal K-10 
#if small use leave-oune-out cross -validation (LOOCV)
scores
```

```python
from numpy import mean 
from numpy import std
print('Accuracy: %.2f SD: (%.2f)'% (mean(scores),std(scores)))
```

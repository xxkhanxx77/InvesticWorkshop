```python
bband = pd.DataFrame(AOTtoSETpb.loc['2018':'2020']).sort_index(ascending=True)
```

```python
bband
```

```python
mean = bband.PBV.ewm(span=100,adjust=False,min_periods=60).mean()
std = bband.PBV.ewm(span=100,adjust=False,min_periods=60).std()
```

```python
short_ema =  bband.PBV.ewm(span=50,adjust=False,min_periods=60).mean()
```

```python
std
```

```python
import matplotlib.pyplot as plt

bband.plot(color='k',figsize=(10,6))
mean.plot(color='r')
short_ema.plot(color='yellow')
plt.fill_between(std.index,mean-2*std,mean+2*std,color='b',alpha=0.2)
plt.fill_between(std.index,mean-1*std,mean+1*std,color='g',alpha=0.2)
```

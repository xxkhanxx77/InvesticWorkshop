```python
import pandas as pd
import pandas_ta as ta
import numpy as np
import yfinance as yf
import vectorbt as vbt
from datetime import datetime
```

```python
data= pd.DataFrame()
data=data.ta.ticker('AAPL',period='3y',interval='1d')
```

```python
df=data.copy()
```

```python
df.head()
```

```python
custom_strategy = ta.Strategy(
name = 'just_for_fun',
description = 'just_for_test',
ta=[{'kind':'sma','length':20},
    {'kind':'ema','length':5}])
```

```python
df.ta.strategy(custom_strategy)
```

```python
df.tail()
```

```python
def my_strategy(df):
    if(df.Close<df.EMA_5)and (df.Close <df.SMA_20):
        return True
    else :
        return False
```

```python
df['signal']=df.apply(my_strategy,axis=1).vbt.fshift(1)
df
```

```python
# clear nan at first value
df=df.iloc[1:,:]
df
```

```python
signal_vectorbt=df.ta.tsignals(df.signal,asbool=True,append=True)
signal_vectorbt
```

```python
port = vbt.Portfolio.from_signals(
                                 df.Close,
                                 entries = signal_vectorbt.TS_Entries,
                                 exits = signal_vectorbt.TS_Exits,                                
                                 freq='D',
                                 init_cash = 100000,
                                 fees =0.0025,
                                 slippage = 0.0025,
                                 direction=2)
```

```python
port2 = vbt.Portfolio.from_signals(
                                 df.Close,
                                 entries = signal_vectorbt.TS_Entries,
                                 exits = signal_vectorbt.TS_Exits,                                
                                 freq='D',
                                 init_cash = 100000,
                                 fees =0.0025,
                                 slippage = 0.0025,
                                 direction='all')
```

```python
port.plot().show()
```

```python
custom_strategy2 = ta.Strategy(name = 'just_for_fun2',
                               description = 'just_for_test2',
                               ta = [{'kind':'rsi','length':14},
                                    {'kind':'macd'}])
```

```python
df2 = data.copy()
df2.ta.strategy(custom_strategy2)
df2
```

```python
def my_strategy_buydip_sellhigh(df):
    if(df.RSI_14>20) and (df.RSI_14<80)and (df.MACD_12_26_9 > df.MACDs_12_26_9):
        return True
    else :
        return False
```

```python
df2['signal']=df2.apply(my_strategy_buydip_sellhigh,axis=1)
```

```python
signal_vectorbt2 = df2.ta.tsignals(df2.signal,asbool=True,append=True,trade_offset=1)
# trade offset = shift signal value if live data use 0 if backtest use 1
```

```python
signal_vectorbt2
```

```python
port_buydip_sellhigh = vbt.Portfolio.from_signals(df2.Open,
                                                 entries = signal_vectorbt2.TS_Entries,
                                                 exits = signal_vectorbt2.TS_Exits,
                                                 freq='D',
                                                 init_cash=100000,
                                                 fees=0.0025,
                                                 slippage=0.0025,
                                                 direction=0)
```

```python
port_buydip_sellhigh.plot().show()
```

```python
# create market bet indicator
df4['52-week-high']=df4.groupby('Stock')['Close'].transform('max')
df4['52-week-low']=df4.groupby('Stock')['Close'].transform('min')
```

```python

df4['from-high']=df4['Close']/df4['52-week-high']-1
```

```python
df4
```

```python
df4.loc[(df4['RSI_14']>50)&(df4['MACDh_12_26_9']>0) 
        &(df4['MACD_12_26_9']>0)&(df4['from-high']> -0.2),'Action']='Hodl'

df4.loc[(df4['RSI_14']>50)&(df4['MACDh_12_26_9']<0)&(df4['from-high']> -0.2),'Action']='Aware'
```

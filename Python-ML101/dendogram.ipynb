```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = yf.download('BTC-USD , ETH-USD ,DOGE-USD ,SHIB-USD,LUNC-USD,ADA-USD,SOL-USD , MATIC-USD,AVAX-USD,SPY, TDEX.BK ,GC=F,CL=F,BZ=F',
                  period ='1y',interval = '1d')
```

```python
df=data.Close
```

```python
df=data.Close.resample('B').last().iloc[2:]
```

```python
df
```

```python
df=df.fillna(method='ffill')
```

```python
ret =df.pct_change().iloc[1:]
```

```python
corr=ret.corr()
```

```python
corr
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked = linkage(corr,'ward')
labelList =df.columns.to_list()
plt.figure(figsize=(10,10),alpha=0.8)
dendrogram(linked,
          orientation='top',
          labels=labelList)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked = linkage(corr,'ward')
labelList =df.columns.to_list()
plt.figure(figsize=(10,10),alpha=0.8)
dendrogram(linked,
          orientation='top',
          labels=labelList,
          leaf_rotation=90)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked = linkage(corr,'ward')
labelList =df.columns.to_list()
plt.figure(figsize=(10,10),alpha=0.8)
dendrogram(linked,
          orientation='left',
          labels=labelList,
          leaf_rotation=90)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linked = linkage(corr,'ward')
labelList =df.columns.to_list()
plt.figure(figsize=(10,10),alpha=0.8)
dendrogram(linked,
          orientation='top',
          labels=labelList,
          leaf_rotation=45)
plt.show()
```

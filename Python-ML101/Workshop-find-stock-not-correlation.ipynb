```python
import pandas as pd
import yfinance as yf
import seaborn as sns
```

```python
data = yf.download(" BBL.BK KBANK.BK BAY.BK TTB.BK KKP.BK TISCO.BK SAWAD.BK MTC.BK AMANAH.BK AEONTS.BK ASK.BK CHAYO.BK JMT.BK KTC.BK MICRO.BK MTC.BK NCAP.BK S11.BK SAK.BK THANI.BK TK.BK",
                   start = '2018-01-01',end='2022-10-14',period='1d')
```

```python
df = data['Adj Close'].copy()
```

```python
df.stack().groupby(level=1).head(1)
```

```python
df = df.fillna(method='ffill')
df
```

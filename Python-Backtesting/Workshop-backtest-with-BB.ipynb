```python
bb = ta.bbands(df.Price,length=30,std=2,mamode="ema")
rsi = ta.rsi(df.Price,length=14)
```

```python
bb
```

```python
Long_Entry = (df.Price < bb['BBL_30_2.0']) & (rsi < 30) 
Exit = ( bb['BBM_30_2.0'])
Short_Entry = (df.Price > bb['BBU_30_2.0'])& (rsi>70) 
Long_Exit = (rsi>70) & (df.Price > bb['BBU_30_2.0'])
Short_Exit = (rsi<30) | (df.Price < bb['BBU_30_2.0'])
```

```python
Long_Entry.shift(1)
Long_Exit.shift(1)
Short_Entry.shift(1)
Short_Exit.shift(1)
Exit.shift(1)
Long_Entry.fillna(False,inplace =True)
Long_Exit.fillna(False,inplace=True)
Exit.fillna(False,inplace=True)
Short_Entry.fillna(False,inplace=True)
Short_Exit.fillna(False,inplace=True)
```

```python
Entry_Long_signal_vectorbt = ta.tsignals(Long_Entry,asbool=True,append=True)
Exit_Long_signal_vectorbt = ta.tsignals(Long_Exit,asbool=True,append=True)
Exit_signal_vectorbt = ta.tsignals(Exit,asbool=True,append=True)
Entry_Short_signal_vectorbt = ta.tsignals(Short_Entry,asbool=True,append=True)
Exit_Short_signal_vectorbt = ta.tsignals(Short_Exit,asbool=True,append=True)
```

```python
port=vbt.Portfolio.from_signals(df.Price,
                               entries=Entry_Long_signal_vectorbt.TS_Entries,
                               exits =Exit_Long_signal_vectorbt.TS_Entries,
                               short_entries = Entry_Short_signal_vectorbt.TS_Entries,
                               short_exits=Exit_Short_signal_vectorbt.TS_Entries,
                               upon_opposite_entry = 1,
                               freq='1D',
                               init_cash=10000,
                               fees=0.0025,
                               slippage=0.0025,
                               sl_stop=0.15,
                                
                               
  )
```

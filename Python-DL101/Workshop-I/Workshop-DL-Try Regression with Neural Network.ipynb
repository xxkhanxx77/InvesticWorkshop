```python
import pandas as pd
import yfinance as yf
```

```python
start = '2021-01-01'
data = yf.download('OR.BK PTTGC.BK PTTEP.BK PTT.BK', start=start)
```

```python
df = data.Close.pct_change().dropna()
df
```

```python
df = df[['OR.BK','PTTEP.BK','PTTGC.BK','PTT.BK']]
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
model = Sequential()
model.add(Dense(1,activation=None, input_dim=3))
model.add(Dense(1))
```

```python
print(model.summary())
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model.plot.png', show_shapes=True, show_layer_names=True)
```

```python
conda install python-graphviz
```

```python
from ann_visualizer.visualize import ann_viz
import graphviz

ann_viz(model, view=True, title='Workshop I')

with open('network.gv') as f:
  dot_graph = f.read()

graphviz.Source(dot_graph)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='MeanSquaredError', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=16)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Data.xlsx',engine ='openpyxl')
```

```python
data
```

# x axis = ROE
# Y axis = PBV

# marketcap = Bubble Chart

```python
fig, ax = plt.subplots()

#ใช้ Scatter Plot เพื่อสร้าง Bubble scatter( แกน X ROE, แกน y PBV, S = ขนาดของ Bubble, Alpha เป็นขนาดความเข้มของ Bubble) 
#ใช้ annotate เพื่อใส่ชื่อของหุ้นแต่ละตัว annotate(ชื่อหุ้น, (แกน x ROE, แกน y PBV))
#ทั้งหมดจะ run ใน for loop ของ dataframe

for i, stock in data.iterrows():
    ax.scatter(stock.ROE, stock.PBV, s = stock.MARKETCAP/100, alpha=0.5)
    ax.annotate(stock.Stocks, (stock.ROE, stock.PBV))

#Plot title แกน x แกน y และ Set xlim และ ylim

plt.title('Bubble Chart')
plt.xlabel('ROE')
plt.ylabel('P/BV')
plt.xlim(0,30)
plt.ylim(0,8)
plt.show()
```

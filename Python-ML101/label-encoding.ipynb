```python
df_ret_1d.Close.apply(lambda x: 1 if  x>0.02 else 0)
```

```python
def myrange(ret):
  if ret <= -0.02:
    return -2
  elif (ret > -0.02) & (ret <0):
    return -1
  elif (ret >=0) & (ret <0.02):
    return 1
  elif ret >= 0.02:
    return 2
```

```python
df_ret_1d.Close.apply(myrange)
```

```python
df_ret_1d.Close = df_ret_1d.Close.apply(myrange)
df_ret_1d
```

```python
dfbackup = df.copy()
```

```python
dfbackup = dfbackup.asfreq('D').fillna(method='ffill')
```

```python
dfbackup.columns.levels[0]
```

```python
# Loop to creates indicator values
stock_indi = {}

for ticker in dfbackup.columns.levels[0]:

    
    df = dfbackup[ticker]
    df['EMA_Long'] =   ta.ema(close = df['Close'],length = 50)
    df['EMA_Short']=   ta.ema(close = df['Close'],length = 20)
    df['RSI']=   ta.rsi(close = df['Close'],length = 10)
    df['52w_high']= df['High'].rolling(window=52*7).max()
    df['drop_from_high'] = df['Close'] / df['52w_high'] -1

    ## Condition
    df['Cond_1'] = df['RSI'] >=65.0 
    df['Cond_2'] = df['EMA_Short'] > df['EMA_Long']
    df['Cond_3'] = df['drop_from_high'] >= -0.1
    df['Signal'] =(df['Cond_1']&df['Cond_2']&df['Cond_3'])

    stock_indi[ticker] = df
```

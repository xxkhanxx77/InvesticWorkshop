```python
ktb = data['KTB']
ktb=ktb.set_index('Date')
ktb=ktb.loc['12/2018':]
```

```python
data2 = pd.concat([scb.iloc[:,1:],kbank.iloc[:,1:],bbl.iloc[:,1:],bay.iloc[:,1:],ktb.iloc[:,1:]],axis=1)
```

```python
data2
```

```python
df_pe = data2['P/E']
df_pbv = data2['P/BV']
df_peg =data2['PEG']
```

```python
df_pe_mean = df_pe.mean(axis=1)
df_pbv_mean = df_pbv.mean(axis=1)
df_peg = df_peg.mean(axis=1)
```

```python
df_p = pd.concat([df_pe_mean,df_pbv_mean,df_peg],axis=1,keys=['PE','PBV','PEG'])
```

```python
df_p
```

```python
def pe(pe):
    if pe<=8:
        return 1
    elif (pe >8)&(pe<=9):
        return 2
    elif (pe>9) &(pe<=10):
        return 3
    elif (pe>10)&(pe<=11):
        return 4
    elif pe>11:
        return 5
```

```python
def pbv(pbv):
    if pbv<=0.6:
        return 1
    elif ( pbv > 0.6) &(pbv <=0.7):
        return 2
    elif (pbv> 0.7)&(pbv <=0.8):
        return 3
    elif (pbv>0.8)& (pbv<=0.9):
        return 4
    elif pbv>0.9:    
        return 5
```

```python
def peg(peg):
    if peg <= 0:
        return -1
    elif (peg > 0) & (peg <= 1):
        return 1
    elif (peg > 1) & (peg <= 2):
        return 2
    elif (peg > 2) & (peg <= 3):
        return 3
    elif (peg > 4) & (peg <= 5):
        return 4
    elif peg > 5:
        return 5
```

```python
df_p['PE_Label']=df_p['PE'].apply(pe)
df_p['PBV_Label']=df_p['PBV'].apply(pbv)
df_p['PEG_Label']=df_p['PEG'].apply(peg)
```

```python
df_p
```

```python
df = pd.concat([df_p,scb.iloc[:,4]], axis=1)
```

```python
df['Return_Shift'] = df.Close.pct_change().shift(-12)
```

```python
df.dropna(inplace=True)
```

```python
df = df[['PE','PBV','PEG','PE_Label','PBV_Label','PEG_Label','Return_Shift']]
```

```python
df
```

```python
X = df.iloc[:14, :-1]
y = df.iloc[:14, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

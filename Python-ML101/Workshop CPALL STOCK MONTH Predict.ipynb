```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('CPALL_Month.xlsx', index_col=0)
```

```python
data
```

```python
data= data.loc['12/2018':]
```

```python
df=data[['Close','P/E','P/BV','PEG']]
```

```python
df
```

```python
def pe(pe):
    if pe <=26:
        return 1
    elif (pe>26)&(pe<=28):
        return 2
    elif (pe>28)&(pe<=30):
        return 3
    elif (pe > 30) & (pe <= 34):
        return 4
    elif (pe > 34) & (pe <= 38):
        return 5
    elif (pe > 38) & (pe <= 42):
        return 6
    elif pe > 42:
        return 7
```

```python
def pbv(pbv):
    if pbv <= 5.8:
        return 1
    elif (pbv > 5.8) & (pbv <= 6.4):
        return 2
    elif (pbv > 6.4) & (pbv <= 7.0):
        return 3
    elif (pbv > 7.0) & (pbv <= 7.6):
        return 4
    elif (pbv > 7.6) & (pbv <= 8.2):
        return 5
    elif (pbv > 8.2) & (pbv <= 9.0):
        return 6
    elif pbv > 9.0:
        return 7
```

```python
def peg(peg):
    if peg <= -2:
        return -2
    elif (peg > -2) & (peg <= 0):
        return -1
    elif (peg > 0) & (peg <= 2):
        return 1
    elif (peg > 2) & (peg <= 4):
        return 2
    elif (peg > 4) & (peg <= 6):
        return 3
    elif (peg > 6) & (peg <= 8):
        return 4
    elif (peg > 8) & (peg <= 10):
        return 5
    elif (peg > 10) & (peg <= 12):
        return 6
    elif peg > 12:
        return 7
```

```python
df['PE_Label'] = df['P/E'].apply(pe)
df['PBV_Label'] = df['P/BV'].apply(pbv)
df['PEG_Label'] = df['PEG'].apply(peg)
```

```python
df['Return'] = df.Close.pct_change()
```

```python
df
```

```python
start_remove = '03/2020'
end_remove = '07/2020'
```

```python
df = pd.concat([df[:start_remove], df[end_remove:]])
```

```python
df['Return_Shift'] = df.Return.shift(-12)
```

```python
df = df[['P/E','P/BV','PEG','PE_Label','PBV_Label','PEG_Label','Return_Shift']]
df.dropna(inplace=True)
```

#  if marketcap come with string format can use this

```python
df['Market Cap']= df['Market Cap'].apply(lambda x: x.replace(',', ''))
```

```python
df['Market Cap']=df['Market Cap'].apply(lambda x: float(x))
```

```python
df
```

```python
df['Label'] = df['Market Cap'].apply(lambda x: 3 if x > 1e10 else (2 if 2e9 < x < 1e10 else (1 if 3e8 < x <2e9 else 0)))
```

# alternative way

```python
df['Label2']=df['Market Cap'].apply(lambda x :3 if x >  10000000000 else (2 if 2000000000 < x < 10000000000 else(1 if 30000000 <x <2000000000 else 0)))  
```

```python
import pandas as pd
import yfinance as yf
```

```python
start = '2021-01-01'
end = '2022-01-01'
```

```python
data_price_condition = yf.download('BTC-USD', start='2021-01-01', end='2022-05-01')
```

```python
df_price_condition = data_price_condition.copy()
```

```python
df_price_condition['Return']= df_price_condition['Close'].pct_change()
```

```python
df_price_condition
```

```python
df_price_condition[df_price_condition['Close']>60000]
```

```python
df_price_condition[df_price_condition.Return>0]
```

```python
def checkGainLoss(df_return):
    if df_return > 0:
        return 'Gain'
    elif pd.isna(df_return):
        return 'Unknown'
    else:
        return 'Loss'
```

```python
df_price_condition['result']=df_price_condition.Return.map(checkGainLoss)
```

```python
df_price_condition.dropna(inplace=True)
```
